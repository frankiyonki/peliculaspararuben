Se realizaron encuestas para recopilar las opiniones de los espectadores sobre una película, calificándola del 0 al 5. A partir de los datos obtenidos, se calculó la media de las opiniones, que resultó ser de 2.51. Además, se determinó que la varianza de las opiniones fue de 1.82 y la desviación estándar fue de 1.35. Se pide analizar la distribución de las opiniones en relación a la media y la desviación estándar, para determinar el porcentaje de observaciones comprendidas dentro de los intervalos de 1 desviación estándar, 2 desviaciones estándar y 3 desviaciones estándar respecto a la media

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Creación del dataset:

```python
opiniones = [5,4,3,2,1,0]
votantes = [42,96,132,124,88,58]

data = {'Opinión':opiniones, 'Cantidad de votantes': votantes}

df = pd.DataFrame(data)
```

Visualización gráfica de los datos

```python
plt.bar(opiniones, votantes)

plt.plot(opiniones, votantes, color='red', linewidth=2.5)

plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)

plt.gca().invert_xaxis()

plt.title('Opiniones obtenidas para una película')
plt.xlabel('Opinión')
plt.ylabel('Cantidad de votantes')

plt.show()
```

Cálculos:


Media:

```python
df['Productos'] = df['Opinión'] * df['Cantidad de votantes']

suma_productos = df['Productos'].sum()
suma_frecuencias = df['Cantidad de votantes'].sum()

media = suma_productos / suma_frecuencias

print(media)
```

Varianza:


```python
df['Diferencia al cuadrado'] = (df['Opinión'] - media) ** 2

df['Producto frecuencia * diferencia al cuadrado'] = df['Cantidad de votantes'] * df['Diferencia al cuadrado']

varianza = df['Producto frecuencia * diferencia al cuadrado'].sum() / suma_frecuencias

print(varianza)
```

Desviación típica

```python
desviacion_típica = np.sqrt(varianza)

print(desviacion_típica)
```

Porcentages de observaciones por cada desviación

```python
observaciones_1_desviacion = df[df['Opinión'].between(media - desviacion_típica, media + desviacion_típica)]['Cantidad de votantes'].sum()
observaciones_2_desviaciones = df[df['Opinión'].between(media - 2*desviacion_típica, media + 2*desviacion_típica)]['Cantidad de votantes'].sum()
observaciones_3_desviaciones = df[df['Opinión'].between(media - 3*desviacion_típica, media + 3*desviacion_típica)]['Cantidad de votantes'].sum()

porcentaje_1_desviacion = observaciones_1_desviacion / suma_frecuencias * 100
porcentaje_2_desviaciones = observaciones_2_desviaciones / suma_frecuencias * 100
porcentaje_3_desviaciones = observaciones_3_desviaciones / suma_frecuencias * 100
```

Todos los resultados:

```python
print("Media:", media)
print("Varianza:", varianza)
print("Desviación estándar:", desviacion_típica)
print("Porcentaje de observaciones dentro de 1 desviación típicas:", porcentaje_1_desviacion)
print("Porcentaje de observaciones dentro de 2 desviaciones típicas:", porcentaje_2_desviaciones)
print("Porcentaje de observaciones dentro de 3 desviaciones típicas:", porcentaje_3_desviaciones)
```
